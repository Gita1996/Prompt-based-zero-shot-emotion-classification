```python
import argparse
import statistics
import numpy as np
import pandas as pd

import torch
from transformers import AutoModelForSequenceClassification, AutoTokenizer

from sklearn.metrics import classification_report

import statistics
from scipy.stats import entropy

#Isear_nli_syn_avg_emo_s_prompt.csv
#data = pd.read_csv('Isear_nli_syn_avg_emotion_prompt2.csv')
#data=pd.read_csv('Isear_nli_syn_avg_emotion_prompt2.csv')
#data=pd.read_csv('Isear_nli_syn_max_emotion_prompt.csv')
#Isear_nli_syn_avg_emotion_prompt2
#TEC_nli_syn_avg_emotion_prompt
data=pd.read_csv('TEC_nli_syn_avg_emotion_prompt.csv')

l=data['labels'].tolist()
e=data['emotion'].tolist()
anger=0
disgust=0
fear=0
joy=0
sadness=0
surprise=0

for i in range(len(l)):
    if l[i]=='sadness':
        if e[i]=='anger':
            anger+=1
        if e[i]=='disgust':
            disgust+=1
        if e[i]=='fear':
            fear+=1
        if e[i]=='joy':
            joy+=1
        if e[i]=='sadness':
            sadness+=1
        if e[i]=='surprise':
            surprise+=1
print(anger)
print(disgust)
print(fear)
print(joy)
print(sadness)
print(surprise)
```
